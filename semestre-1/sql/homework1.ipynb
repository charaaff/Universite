```python
%%sql
DROP TABLE IF EXISTS PASSENGER;
DROP TABLE IF EXISTS TICKET;

CREATE TABLE TICKET(
    PassengerId integer,
    Ticket varchar(20),
    Fare float,
    Cabin varchar(20),
    Embarked varchar(10),
    PRIMARY KEY(PassengerId)
);

CREATE TABLE PASSENGER(
    PassengerId integer, 
    Survived integer,
    Pclass integer,
    Name varchar(100), 
    Sex varchar(10), 
    Age float, 
    SibSp integer,
    Parch integer,
    PRIMARY KEY(Name)
);
```
